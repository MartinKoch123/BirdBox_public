```python
import sys
sys.path.append("..\..")
from birdbox.tools import StopFile, fancy_loop
from birdbox.database import SearchTerm, Search, Url, Bird, SearchHit
from birdbox.scraping import search, download, search_macaulay
from progressbar import progressbar
from peewee import JOIN
DOWNLOAD_FOLDER = r"images\database"
```

## Scrape web for image URLs

```python
query = SearchTerm.select().where(SearchTerm.source == 1).join(Search, JOIN.LEFT_OUTER).where(Search.search_term.is_null())
for e in query:
    print(e)
```

```python
query = SearchTerm.select().where(SearchTerm.bird == Bird.find("wood"), SearchTerm.source == 1)
```

```python
for search_term_entry in fancy_loop(query.limit(1)):
    search(search_term_entry)
```

```python
for search_term_entry in fancy_loop(query):
    search_macaulay(search_term_entry)
```

## Download images

```python
query = Url.select().where(Url.image.is_null(), Url.is_broken == 0)

for url_entry in fancy_loop(query):
    download(DOWNLOAD_FOLDER, url_entry)
```

## Overview

```python
for bird_entry in Bird.select():
    url_count = (
        Url
        .select()
        .join(SearchHit)
        .join(SearchTerm)
        .where(SearchTerm.bird == bird_entry)
        .count()
    )
    print(f"{bird_entry.nickname()}: {url_count}")
```
